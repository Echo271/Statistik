```
#import library yang diperlukan

import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline
import matplotlib.pyplot as plt
from collections import OrderedDict
```

```
#open dataSet yang telah di download dari MOLS 
data = pd.read_csv('/content/Dr.-Fahrul-Agus,MT_CovidTI_A_DataSet (1).csv')

#mengecek dataset dengan .head sehingga menampilkan dataset paling atas
data.head()
```

```
kurunWaktu = data['Bulan']

#
from collections import OrderedDict
kurunWaktu=list(OrderedDict.fromkeys(kurunWaktu))

#
timeline = [3,4,5,6,7,8]
data=data.replace(kurunWaktu, timeline)
```

```
#
bpn=data.loc[data['Wilayah']=="BPN"]
ppu=data.loc[data['Wilayah']=="PPU"]
paser=data.loc[data['Wilayah']=="PASER"]
kubar=data.loc[data['Wilayah']=="KUBAR"]
```

```
#Balikpapan
plt.subplots(figsize=(15, 6))
sumbu = bpn
plt.title("Kasus Covid-19 di Balikpapan dari Bulan Maret-Agustus 2020")
sns.lineplot(sumbu['Bulan'], sumbu['Total Kasus Positif Terkonfirmasi'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien dalam Perawatan'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Sembuh'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Meninggal'])
plt.ylabel('Jumlah Kasus')
plt.xlabel('Bulan')
plt.legend(['Kasus Positif', 'Dirawat', "Sembuh", 'Meninggal'], loc='upper left')
plt.show()
```

```
#PPU
plt.subplots(figsize=(15, 6))
sumbu = ppu
plt.title("Kasus Covid-19 di Kab.PPU dari Bulan Maret-Agustus 2020")
sns.lineplot(sumbu['Bulan'], sumbu['Total Kasus Positif Terkonfirmasi'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien dalam Perawatan'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Sembuh'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Meninggal'])
plt.ylabel('Jumlah Kasus')
plt.xlabel('Bulan')
plt.legend(['Kasus Positif', 'Dirawat', "Sembuh", 'Meninggal'], loc='upper left')
plt.show()
```

```
#Paser
plt.subplots(figsize=(15, 6))
sumbu = paser
plt.title("Kasus Covid-19 di Kab.Paser dari Bulan Maret-Agustus 2020")
sns.lineplot(sumbu['Bulan'], sumbu['Total Kasus Positif Terkonfirmasi'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien dalam Perawatan'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Sembuh'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Meninggal'])
plt.ylabel('Jumlah Kasus')
plt.xlabel('Bulan')
plt.legend(['Kasus Positif', 'Dirawat', "Sembuh", 'Meninggal'], loc='upper left')
plt.show()
```

```
#KUBAR
plt.subplots(figsize=(15, 6))
sumbu = kubar
plt.title("Kasus Covid-19 di Kab.Kutai Barat dari Bulan Maret-Agustus 2020")
plt.ylabel('Jumlah Kasus')
plt.xlabel('Bulan')
plt.legend(['Kasus Positif', 'Dirawat', "Sembuh", 'Meninggal'], loc='upper left')
sns.lineplot(sumbu['Bulan'], sumbu['Total Kasus Positif Terkonfirmasi'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien dalam Perawatan'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Sembuh'])
sns.lineplot(sumbu['Bulan'], sumbu['Total Pasien Meninggal'])
plt.show()
```
